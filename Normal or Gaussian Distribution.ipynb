```python
import numpy
from numpy import random 
import matplotlib.pyplot as plt #for making graphs
import seaborn as sns #for beatufication of graph
```

```python
x=random.normal(size=1000)
print(x)
```

```python
x=random.normal(size=(2,3))
print(x)
```

```python
x=random.normal(loc=1,scale=2,size=(2,3))
print(x)
```

```python
sns.histplot(random.normal(size=1000))
plt.show()
```

```python
sns.histplot(random.binomial(n=10,p=.5,size=1000), kde=True)
plt.show()
```

```python
x=random.poisson(lam=2,size=10)
print(x)
```

```python
sns.distplot(random.poisson(lam=2,size=1000), hist=True, kde=True)
plt.show()
```
